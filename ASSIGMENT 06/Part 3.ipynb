# *Part 3 (Array Creation Functions)*

```python
import numpy as np
```

```python
# a)    Part 1
zero_matrix = np.zeros((4, 4))
print("Zero Matrix (4x4):")
print(zero_matrix)
```

```python
# b)
one_matrix = np.ones((3, 2))
print("One Matrix (3x2):")
print(one_matrix)
```

```python
# c)
identity_matrix = np.eye(5)
print("Identity Matrix (5x5):")
print(identity_matrix)
```

```python
# d)
constant_array = np.full((3, 3), 7)
print("Constant Array (3x3) filled with 7:")
print(constant_array)
```

```python
# e)
random_array = np.random.randint(10, 100, size=(3, 4))
print("Random Integer Array (3x4) between 10 and 99:")
print(random_array)
```

```python
# Part 2
arange_array = np.arange(5, 50, 5).reshape(3, 3)
print("Array from 5 to 50 (step 5), reshaped to (3x3):")
print(arange_array)
```
