Task 2 (Student Marks Validation)

```python
class Student:
    def __init__(self, name, roll_no, marks=0):
        # Private attributes
        self.__name = name
        self.__roll_no = roll_no
        self.__marks = 0  
        self.set_marks(marks)  

    # Setter method for marks (accepts 0â€“100)
    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks! Please enter a value between 0 and 100.")

    # Getter methods
    def get_name(self):
        return self.__name

    def get_roll_no(self):
        return self.__roll_no

    def get_marks(self):
        return self.__marks

```

```python
student1 = Student("Farjad", 301, 75)
```

```python
print("name:", student1.get_name())
print("Roll no:", student1.get_roll_no())
print("Marks:", student1.get_marks())
```

```python
student1.set_marks(120)
```
