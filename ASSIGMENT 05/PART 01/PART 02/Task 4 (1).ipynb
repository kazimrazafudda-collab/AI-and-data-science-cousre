```python
# Parent class 1
class Father:
    def skills(self):
        return "Scientist"

# Parent class 2
class Mother:
    def skills(self):
        return "Freelancer"

# Child class Father and Mother
class Child(Father, Mother):
    def skills(self):
        # Combine both parents' skills
        father_skill = Father.skills(self)
        mother_skill = Mother.skills(self)
        return f"{father_skill} and {mother_skill}"
```

```python
child1 = Child()
```

```python
print("childskills: ", child1.skills())
```
