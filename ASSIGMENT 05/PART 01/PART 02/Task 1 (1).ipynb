```python
# Parent class
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

# Child class (inherits from Animal)
class Dog(Animal):
    # Method overriding
    def make_sound(self):
        print("Bark!")

```

```python
a = Animal()
d = Dog()
```

```python
a.make_sound()
d.make_sound()
```
